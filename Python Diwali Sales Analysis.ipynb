# Exploartary Data Analysis (EDA)

```python
ax=sns.countplot(x='Gender',data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Gender_Sales=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=Gender_Sales)
```

From the above graphs it is concluded that most of the buyers are females, even the purchasing power of women of females are greater than males.

# Age

```python
age=sns.countplot(x='Age Group',data=data,hue='Gender')
for bar in age.containers:
    age.bar_label(bar)
```

```python
Age_Sales=data.groupby(['Age Group','Gender'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Age_Sales,hue='Gender')
```

From the above graphs it is concluded that most of the buyers are the females of age group between 26-35 years.

# State

```python
State_Orders=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set({'figure.figsize':(15,4)})
order_state=sns.barplot(x='State',y='Orders',data=State_Orders)
for bar in order_state.containers:
    order_state.bar_label(bar)
```

```python
State_Sales=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set({'figure.figsize':(15,4)})
sns.barplot(x='State',y='Amount',data=State_Sales)
```

From the above graphs it is concluded that most of the orders and sales are from Uttar Pradesh, Maharashtra and Karnataka, respectively.

# Martial Status

```python
count_martial_status=sns.countplot(x='Marital_Status',data=data)
sns.set({"figure.figsize":(5,2)})
for bar in count_martial_status.containers:
    count_martial_status.bar_label(bar)
```

```python
Sales_Martial_Status=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set({"figure.figsize":(3,4)})
sns.barplot(x='Marital_Status',y='Amount',data=Sales_Martial_Status,hue='Gender')
```

From the above graphs it is concluded that most of the buyers are married females and they have more purchasing power. 

# Occupation

```python
sns.set({"figure.figsize":(20,5)})
count_occupation=sns.countplot(x='Occupation',data=data)
for bar in count_occupation.containers:
    count_occupation.bar_label(bar)
```

```python
sns.set({"figure.figsize":(20,5)})
Sales_Occupation=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
sns.barplot(x='Occupation',y='Amount',data=Sales_Occupation)
```

From above graphs we can conclude that most of the buyers are working in IT, Healthcare and Aviation sector.

# Product Category

```python
sns.set({"figure.figsize":(22,5)})
count_Product_Category=sns.countplot(x='Product_Category',data=data)
for bar in count_Product_Category.containers:
    count_Product_Category.bar_label(bar)
```

```python
sns.set({"figure.figsize":(20,5)})
sales_Product_Category=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=sales_Product_Category)
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category.

```python
sns.set({"figure.figsize":(22,5)})
sales_Product=data.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_Product)
```

# Conclusion

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
